```python
import seaborn as sns
%matplotlib inline
tips= sns.load_dataset('tips')
tips.head()
```

```python
import numpy as np
```

```python
sns.barplot(x= 'sex' ,y= 'total_bill' ,data = tips, estimator = np.std )
```

```python
sns.countplot(x= 'sex',data = tips)
```

```python
tips
```

```python
sns.boxplot(x = 'day',y= 'total_bill', data = tips)
```

```python
sns.boxplot(x = 'day',y= 'total_bill', data = tips, hue = 'smoker')
```

```python
sns.violinplot(x = 'day',y= 'total_bill', data = tips)
```

```python
sns.boxplot(x = 'day',y= 'total_bill', data = tips)
```

```python
sns.violinplot(x = 'day',y= 'total_bill', data = tips, hue='sex')
```

```python
sns.violinplot(x = 'day',y= 'total_bill', data = tips, hue='sex', split=True)
```

```python
sns.stripplot(x = 'day', y = 'total_bill', data = tips, jitter = True)
```

```python
sns.stripplot(x = 'day', y = 'total_bill', data = tips, jitter = False)
```

```python
sns.stripplot(x = 'day', y = 'total_bill', data = tips, hue ='sex')
```

```python
sns.stripplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex', dodge = True)
```

```python
# violin plot + strip plot = SWARM PLOT
sns.swarmplot(x='day', y='total_bill', data = tips)
#swarm plot sometimes do not scale too well on very large data sets
#avoid them because most peole do not know about them
```

```python
#better method 
sns.violinplot(x='day', y='total_bill', data = tips)
sns.swarmplot(x='day', y='total_bill', data = tips, color='black')
```

```python
#factor plot (general method)
sns.factorplot(x='day',y='total_bill',data=tips,kind='bar')
```

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='violin')
```
