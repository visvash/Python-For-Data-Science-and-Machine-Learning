```python
import numpy as np
import matplotlib.pyplot as plt   #for visualization
%matplotlib inline                          
#matplotlib inline to see inline for everything
```

```python
points = np.arange(-5,5,0.1)      #create points
```

```python
#we will create a grid now points*points
dx,dy = np.meshgrid(points,points)
```

```python
dx               #among rows
```

```python
dy               #among columns
```

```python
z = (np.sin(dx)+np.sin(dy))
#universal function
z
```

```python
#to plot imshow means show as image
plt.imshow(z)
```

```python
plt.imshow(z)
plt.colorbar()

plt.title('Plot for sin(x)+sin(y)')
```

```python
#numpy where

A = np.array([1,2,3,4])

B = np.array([100,200,300,400])
```

```python
condition = np.array([True,True,False,False])
```

```python
#lets say we want ot make an array we choose A value where the condtition is ture otherwise we choose B value
```

```python
answer = [(A_val if cond else B_val) for A_val,B_val,cond in zip(A,B,condition)]
```

#answer will say heyy show me the A val if the conditionn is met else show B value out of each value from the A,B,Condition array

```python
answer
```

```python
answer2 = np.where(condition,A,B) #shortcut for line16
```

```python
answer2
```
