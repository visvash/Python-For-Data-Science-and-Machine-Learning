```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
#functional method
```

```python
plt.plot(x,y,'r--')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#object oriented method
```

```python
fig= plt.figure()
axes = fig.add_axes([.1,.1,.8,.8])

axes.plot(x,y)
axes.set_xlabel('x label')
axes.set_ylabel('y label')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([.1,.1,.8,.8])   #[left,bottom,width,height] in %age
axes2 = fig.add_axes([.2,.5,.4,.3])   #[left,bottom,width,height] in %age
axes1.plot(x,y)
axes1.set_title('Larger plot')
axes2.plot(y,x)
axes2.set_title('Smaller plot')
```

```python
fig,axes = plt.subplots(nrows =1, ncols = 2)

axes[0].plot(x,y)
axes[0].set_title('first plot')

axes[1].plot(y,x)
axes[1].set_title('second plot')

plt.tight_layout()    #use it always
```

<h1>Figure Size and DPI</h1>

```python
fig = plt.figure(figsize = (3,2), dpi = 100) #(dpi = dots/pixels per inch)
ax= fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
#using subplots
fig,axes = plt.subplots(nrows = 2, ncols = 1,figsize = (8,2), dpi = 100) #(dpi = dots/pixels per inch)
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
#legends
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label = 'X squared',)
ax.plot(x,x**3, label = 'X cubed')

ax.legend(loc =0)
```

<h1>PLOT APPEARANCE</h1>

```python
fig = plt.figure()
ax= fig.add_axes([0,0,1,1])
ax.plot(x,y,color = '#fa6b9d',linewidth = 3, alpha = 10)   #linewidth =lw    
#rgb hexcode tooo or just color name for basic colours
```

```python
fig = plt.figure()
ax= fig.add_axes([0,0,1,1])
ax.plot(x,y,color = '#fa6b9d', linewidth=2, ls= '-.')
```

```python
#markers
fig = plt.figure()
ax= fig.add_axes([0,0,1,1])
ax.plot(x,y,color = '#fa6b9d', linewidth=3, ls= '-',
        marker = 'o', markersize =5, markerfacecolor = 'yellow',
       markeredgewidth = 1, markeredgecolor = 'black')
# o + * 1
```

<h1>CONTROL OVER AXIS APPEARANCE AND PLOT RANGE</h1>

```python
fig = plt.figure()
ax= fig.add_axes([0,0,1,1])
ax.plot(x,y,color = '#fa6b9d', linewidth=3, ls= '--')
ax.set_xlim([0,1])
ax.set_ylim([0,4])
```
