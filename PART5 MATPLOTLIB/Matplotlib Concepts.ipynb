```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
#functional method
```

```python
plt.plot(x,y,'r--')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#object oriented method
```

```python
fig= plt.figure()
axes = fig.add_axes([.1,.1,.8,.8])

axes.plot(x,y)
axes.set_xlabel('x label')
axes.set_ylabel('y label')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([.1,.1,.8,.8])   #[left,bottom,width,height] in %age
axes2 = fig.add_axes([.2,.5,.4,.3])   #[left,bottom,width,height] in %age
axes1.plot(x,y)
axes1.set_title('Larger plot')
axes2.plot(y,x)
axes2.set_title('Smaller plot')
```
